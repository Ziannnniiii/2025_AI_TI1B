# 2. Membuat dan Mengakses Elemen Series

# Membuat Series dari sebuah List Python

```
import pandas as pd

li = [1, 7, 2]
se = pd.Series(li)

print(se)
```

# Mengakses Elemen pada Series

```
li = [1, 7, 2]
se = pd.Series(li)

print(se[0])
```

# Memberi Label pada Series

```
li = [1, 7, 2]
se = pd.Series(li, index = ['x', 'y', 'z'])

print(se['x'])
```

# Membuat Series dari Dictionary

```
calories = {"day1": 420, "day2": 380, "day3": 390}
se = pd.Series(calories)

print(se)
```

```
df.iloc[0:2, [0,1]]
df.iloc[0:2, 0:2]
df.iloc[:, 0:2]
df.loc[0:2, :'calories']
df.loc[0:2, ['calories', 'duration']]
df['calories']
df['calories'][0]
df[['calories']]
df[['calories']].loc[[0, 1]]
```
