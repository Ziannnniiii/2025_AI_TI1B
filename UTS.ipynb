```
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
df = pd.read_csv('/content/airbnb.csv')

# Tampilkan 5 data pertama
print("Head DataFrame:\n", df.head())

# Series: Ambil kolom tertentu
print("\nSeries - Nama Host:\n", df['host_name'].head())

# Info dan ringkasan statistik
print("\nInfo DataFrame:")
df_info = df.info()
print(df_info)

print("\nDeskripsi Statistik:")
print(df.describe())

# Manipulasi: Filter data berdasarkan harga < 100
df_filtered = df[df['price'] < 100]
print("\nData dengan harga < 100:\n", df_filtered[['name', 'price']].head())

# Plotting sederhana: Histogram harga
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Jumlah Listing')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ====================================
# 1. Load Data
# ====================================
df = pd.read_csv('/content/airbnb.csv')

# Set gaya visual Seaborn
sns.set(style="whitegrid")

# ====================================
# 2. Histogram Dasar: Distribusi Harga
# ====================================
plt.figure(figsize=(10, 5))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribusi Harga Airbnb')
plt.xlabel('Harga ($)')
plt.ylabel('Jumlah Listing')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# ====================================
# 3. Subplot: Minimum Nights & Jumlah Review
# ====================================
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].hist(df['minimum_nights'], bins=30, color='salmon', edgecolor='black')
axes[0].set_title('Minimum Nights')
axes[0].set_xlabel('Jumlah Malam')
axes[0].set_ylabel('Jumlah Listing')
axes[0].grid(True, linestyle='--', alpha=0.6)

axes[1].hist(df['number_of_reviews'], bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Number of Reviews')
axes[1].set_xlabel('Jumlah Review')
axes[1].set_ylabel('Jumlah Listing')
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# ====================================
# 4. Boxplot Harga per Neighbourhood Group (tanpa filter harga)
# ====================================
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='neighbourhood_group', y='price')
plt.title('Boxplot Harga berdasarkan Wilayah (Tanpa Filter)')
plt.xlabel('Wilayah')
plt.ylabel('Harga ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ====================================
# 5. Histogram Harga dengan KDE & Batas Maksimal
# ====================================
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=100, kde=True)
plt.xlim(0, 1000)
plt.title("Distribusi Harga Airbnb di NYC (dibatasi sampai $1000)")
plt.xlabel("Harga ($)")
plt.ylabel("Jumlah Listing")
plt.tight_layout()
plt.show()

# ====================================
# 6. Boxplot Harga per Neighbourhood Group (dengan filter < $1000)
# ====================================
plt.figure(figsize=(10, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=df[df['price'] < 1000])
plt.title("Harga Airbnb Berdasarkan Wilayah (dibatasi < $1000)")
plt.xlabel("Wilayah")
plt.ylabel("Harga ($)")
plt.tight_layout()
plt.show()

# ====================================
# 7. Boxplot Harga per Room Type (dengan filter < $1000)
# ====================================
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df[df['price'] < 1000])
plt.title("Harga Airbnb Berdasarkan Jenis Kamar (dibatasi < $1000)")
plt.xlabel("Jenis Kamar")
plt.ylabel("Harga ($)")
plt.tight_layout()
plt.show()

# ====================================
# 8. Barplot Jumlah Listing per Wilayah
# ====================================
plt.figure(figsize=(8, 5))
sns.barplot(x=df['neighbourhood_group'].value_counts().index,
            y=df['neighbourhood_group'].value_counts().values,
            palette='pastel')
plt.title("Jumlah Listing per Wilayah")
plt.xlabel("Wilayah")
plt.ylabel("Jumlah Listing")
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ===============================
# 1. Load Data
# ===============================
df = pd.read_csv('/content/airbnb.csv')

# ===============================
# 2. Statistik Deskriptif
# ===============================
print(" Statistik Deskriptif:")
print(df.describe())

print("\n Median Kolom Numerik:")
print(df.median(numeric_only=True))

harga_min = df['price'].min()
harga_max = df['price'].max()
mean_price = df['price'].mean()
std_price = df['price'].std()
perbandingan = harga_max / (harga_min + 1)

print(f"\n Harga Minimum: {harga_min}")
print(f" Harga Maksimum: {harga_max}")
print(f" Rata-rata Harga: {mean_price:.2f}")
print(f" Standar Deviasi Harga: {std_price:.2f}")
print(f" Perbandingan Harga Maks / Min: {perbandingan:.2f}")

# ===============================
# 3. Heatmap Korelasi Numerik
# ===============================
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Korelasi antar Kolom Numerik')
plt.tight_layout()
plt.show()

# ===============================
# 4. Rata-rata Harga per Wilayah
# ===============================
mean_price_per_area = df.groupby('neighbourhood_group')['price'].mean()
print("\n Rata-rata Harga per Wilayah:\n", mean_price_per_area)

# ===============================
# 5. Wilayah dengan Listing Terbanyak
# ===============================
top_neigh = df['neighbourhood'].value_counts().head(5)
print("\n 5 Wilayah dengan Listing Terbanyak:\n", top_neigh)

# ===============================
# 6. Total Review per Wilayah
# ===============================
total_reviews = df.groupby('neighbourhood_group')['number_of_reviews'].sum()
print("\n Total Review per Wilayah:\n", total_reviews)

# ===============================
# 7. Listing Paling Mahal
# ===============================
most_expensive = df.sort_values(by='price', ascending=False).head(1)
print("\n Listing Paling Mahal:\n", most_expensive[['name', 'host_name', 'price']])
```
